```python
import pandas as pd
import math
import numpy as np
```

```python
housingPrices = pd.read_csv("kc_house_data.csv")
```

```python
housingPrices
```

```python
housingPrices.columns
```

```python
housingPrices.drop(columns=['date'],inplace = True)
```

```python
housingPrices.corr()['price']
```

```python
housingPrices = housingPrices[['id','price','bedrooms','bathrooms','sqft_living','grade','sqft_above','sqft_living15']]
```

```python
housingPrices
```

```python
housingPrices[housingPrices.isnull().any(axis = 1)]
```

```python
housingPrices = housingPrices.dropna()
```

```python
train = housingPrices[housingPrices['id']<7715801031].copy()
test = housingPrices[housingPrices['id']>7715801030].copy()
```

```python
train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
predictors = ['bedrooms', 'bathrooms','sqft_living','grade','sqft_above','sqft_living15']
target = 'price'
```

```python
reg = LinearRegression()
```

```python
reg.fit(train[predictors], train['price'])
```

```python
predictions = reg.predict(test[predictors])
```

```python
predictions
```

```python
test['predictions'] = test[predictions].round()
```
